Removing Outliers

```python
from scipy.stats import zscore
z=np.abs(zscore(x))
```

```python
threshold=3
print(np.where (z>3))
```

```python
df_new=x[(z<3).all(axis=1)]
```

# FEATURE ENGINEERING AND MODELLING

The Data doesn't have the Y data (Target) so we will do Unsupervised Learning here.

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

ELBOW METHOD TO KNOW ABOUT CLUSTER

```python
wcss=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df_new)
    wcss.append(km.inertia_)
    
plt.plot(K,wcss,marker='x')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
model=KMeans(n_clusters=4)
y_model=model.fit_predict(df_new)
y_model
```

```python
model.labels_
```

```python
plt.scatter(df_new.iloc[:,0],df_new.iloc[:,1],
           c=model.labels_,cmap='viridis')
```
